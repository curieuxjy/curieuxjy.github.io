## torch.tensor()

```python
import torch
```

```python
original_data = torch.tensor([1])
new_data = torch.tensor(original_data)
print(f"original : {original_data} {original_data.dtype}")
print(f"new : {new_data} {new_data.dtype}")

 
# original dataÎ•º ÏàòÏ†ï
original_data[0] = 2
print(f"original : {original_data} {original_data.dtype}")
print(f"new : {new_data} {new_data.dtype}")
```

- intÏóêÏÑú intÎ°ú
- call by value, ÏÉàÎ°úÏö¥ Tensor Í∞ùÏ≤¥Ïù¥ÎØÄÎ°ú Î≥ÄÌôî x

## torch.Tensor()

```python
original_data = torch.Tensor([1])
new_data = torch.Tensor(original_data)
print(f"original : {original_data} {original_data.dtype}")
print(f"new : {new_data} {new_data.dtype}")

 
# original dataÎ•º ÏàòÏ†ï
original_data[0] = 2
print(f"original : {original_data} {original_data.dtype}")
print(f"new : {new_data} {new_data.dtype}")
```

- intÏóêÏÑú floatÏúºÎ°ú
- call by reference, shallow copy

```python
original_data = [1]
new_data = torch.Tensor(original_data)
print(f"original : {original_data}")
print(f"new : {new_data} {new_data.dtype}")

# original dataÎ•º ÏàòÏ†ï
original_data[0] = 2
print(f"original : {original_data}")
print(f"new : {new_data} {new_data.dtype}")
```

- intÏóêÏÑú floatÏúºÎ°ú
- call by value, deep copy
